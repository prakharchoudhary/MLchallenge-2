```python
import numpy as np
import pandas as pd
from IPython.display import display
```

```python
'''
Load the dataset
'''
train = pd.read_csv("train.csv")
test = pd.read_csv("test.csv")
print ('The train data has {} rows and {} columns'.format(train.shape[0],train.shape[1]))
print ('The test data has {} rows and {} columns'.format(test.shape[0],test.shape[1]))
```

```python
import time

unix_cols = ['deadline', 'state_changed_at', 'created_at', 'launched_at']

for x in unix_cols:
    train[x] = train[x].apply(lambda i: time.ctime(i))
    test[x] = test[x].apply(lambda i: time.ctime(i))
    
display(train[:5])
# print test[:5]
```

```python
'''
creating features
'''

cols_to_use = ['name', 'desc']
f_lens = ['name_len', 'desc_len']
f_count = ['name_count', 'desc_count']

for i in np.arange(2):
    train[f_lens[i]] = train[cols_to_use[i]].apply(str).apply(len)
    train[f_count[i]] = train[cols_to_use[i]].apply(str).apply(lambda x: len(x.split(' ')))
```

```python
train['keywords_len'] = train['keywords'].apply(str).apply(len)
train['keywords_count'] = train['keywords'].apply(str).apply(lambda x: len(x.split('-')))
```

```python
for i in np.arange(2):
    test[f_lens[i]] = test[cols_to_use[i]].apply(str).apply(len)
    test[f_count[i]] = test[cols_to_use[i]].apply(str).apply(lambda x: len(x.split(' ')))
    
test['keywords_len'] = test['keywords'].apply(str).apply(len)
test['keywords_count'] = test['keywords'].apply(str).apply(lambda x: len(x.split('-')))
```

```python
'''
encoding features
'''

from sklearn.preprocessing import LabelEncoder

cols = ['disable_communication', 'country']
for i in cols:
    le = LabelEncoder()
    le.fit(list(train[i].values) + list(test[i].values))
    train[i] = le.transform(list(train[i]))
    test[i] = le.transform(list(test[i].values))
```

```python
display(test)
```

```python
'''
model training
'''

cols_to_use = ['name_len', 'desc_len',
               'name_count', 'desc_count', 
               'keywords_len', 'keywords_count', 'goal', 'country']

target = train['final_status']

# data for modeling
k_train = train[cols_to_use]
k_test = test[cols_to_use]
```

```python
testing = pd.read_csv("samplesubmission.csv")
test_target = testing['final_status']
```

```python
# #using GradientBoostingClassifier
from sklearn.ensemble import GradientBoostingClassifier
clf = GradientBoostingClassifier(loss='exponential', learning_rate=0.1,
                                n_estimators=100, random_state=3)
clf.fit(k_train, target)
pred = clf.predict(k_test)
clf.score(k_test, test_target)
```
